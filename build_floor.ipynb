```python
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family'] = 'simhei'

data_chaoyang = pd.read_excel('C:\\Users\\雪宝\\Desktop\\bj_house_info第三版（副本）.xlsx')

fig=plt.figure(figsize=(6,6))
low=mid=high=down=no=0
for cengshu in data_chaoyang['bulid_floor']:#从build_floor列读取数据
        if cengshu=='低层':
            low+=1
        elif cengshu=='中层':
            mid+=1
        elif cengshu=='高层':
            high+=1
        elif cengshu=='地下':
            down+=1
        elif cengshu=='未说明':
            no+=1
labels=['低层','中层','高层','地下','未说明']#绘图显示的标签
values=[low,mid,high,down,no]
data=Series(values) #收集数据
ax1=fig.add_subplot(2,2,1)
data.plot(kind='pie',figsize=(6,6),title='朝阳区楼层分布饼状图',fontsize=12,startangle = 60,shadow=False,autopct='%1.1f%%',labels=labels,ax=ax1)
plt.ylabel('floor',fontsize=15)
```
