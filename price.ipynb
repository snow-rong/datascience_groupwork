```python
#导入需要用到的模块
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family'] = 'simhei'

fig = plt.figure(figsize=(22,5))
#读取目标表格文件，并用people代表读取到的表格数据
data = pd.read_excel('C:\\Users\\雪宝\\Desktop\\average-price.xlsx')



#x轴是区，y轴是价格，让直方图排序显示，默认升序
data.sort_values(by='price',inplace=True)
plt.ylabel('每平米房价（元/㎡）',fontsize=18)
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.title('各区二手房每平米单价柱形图',fontsize=20)
plt.bar(data['blocks'], data['price'],color='dodgerblue')
rects = plt.bar(data['blocks'], data['price'])
#旋转X轴标签，让其横向写
plt.xticks(rotation=0)
plt.tick_params(labelsize=17) #刻度字体大小17

for rect in rects:  #rects 是柱子的集合
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width() / 2, height, str(height), size=15, ha='center', va='bottom')
plt.savefig('各区二手房每平米单价柱形图')
plt.show()
```
