```python
#
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family'] = 'simhei'

fig = plt.figure(figsize=(22,5))
#people
data = pd.read_excel('C:\\Users\\\\Desktop\\average-price.xlsx')



#xy
data.sort_values(by='price',inplace=True)
plt.ylabel('/',fontsize=18)
plt.rcParams['font.sans-serif']=['SimHei'] #
plt.title('',fontsize=20)
plt.bar(data['blocks'], data['price'],color='dodgerblue')
rects = plt.bar(data['blocks'], data['price'])
#X
plt.xticks(rotation=0)
plt.tick_params(labelsize=17) #17

for rect in rects:  #rects 
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width() / 2, height, str(height), size=15, ha='center', va='bottom')
plt.savefig('')
plt.show()
```
